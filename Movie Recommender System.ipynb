```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
    #Here ast.literal_eval is used to covert a string of list to list
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['cast'][0]
```

<B>We need to filter out the first 3 dictionaries.</B>

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

<B>Now converting overview(string to list).</B>

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

<B>We need to remove spaces between words.</B>

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

<B>Create a column Tags which will be the concatenation of these columns.</B>

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

<B>Vectorization.</B>

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

<B>Stemming</B>

```python
#[love,loving,loved]
#after stemming
#[love,love,love]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

<B>We will calculate cosine distance between vectors(more the diff less the similarity).</B>

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors)
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0].shape
```

```python
fin=list(enumerate(similarity[0]))
#so that if we sort the values stick with the index.
```

```python
sorted(fin,reverse=True,key=lambda x:x[1])[1:6]
```

<B>Let's Recommend!!!</B>

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
